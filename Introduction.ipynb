## A Brief History and why we need a GNC team

Spaceshot is a **Student Research and Development (SRAD)** based team. Unlike teams that utilize commercial off the shelf avionics, we design our electronics, software, and most importantly, GNC algorithms in house. Members get the opportunity to work on the same GNC problems that are used in industry at places like *SpaceX* or *NASA* or at research powerhouses like *MIT LL* or *Draper*. 

As a rocket goes up in the sky, the team back on the ground often wants to know whats going on up there. Thus, there is a need to collect real flight telemetry (which really just means data streaming). But the real challenge is processing it, interpreting what it means, and using it to make informed decisions. That's where the GNC team comes into play. We develop the algorithms and software that transform raw sensor measurements into meaningful information, such as the rocket's true altitude, velocity, orientation, and overall flight state (which refers to just all of those combined). 

## What is MIDAS?

If you're walking down campus and hear the word **MIDAS**, you should know you're in proximity to a Spaceshot avionics member. **MIDAS**, is simple words, is our in-house developed flight computer; you'll hear about it all the time. 

MIDAS serves as the rocket's primary onboard computer and altimeter. It interfaces with sensors, records flight data, performs real-time computations (which is our role!), and makes important flight decisions (which is also our role!) all onboard the rocket. In simple words, you can think of it as the physical brains of the rocket.  

Many of the software tools you'll work with during your time on the team are designed to interface directly with MIDAS or emulate its behavior during testing and simulation.

Comparison operators evaluate relationships and return `True` or `False`. Here are some examples: 

````python
# Comparison operators
a = 10 
b = 5
print(a > b)      # Greater than; Results in True
print(a < b)      # Less than; Results in False
print(a == b)     # Equal to; Results in False
print(a != b)     # Not equal to; Results in True
print(a >= b)     # Greater than or equal to; Results in True
print(a <= b)     # Less than or equal to; Results in False
````

````python
# Print comparison results as 1 (True) or 0 (False) # we add int for python to return 1 or 0 instead of True or False. 
# This is not needed in other languages
print(int(a > b))   # Greater than
print(int(a < b))   # Less than
print(int(a == b))  # Equal to
print(int(a != b))  # Not equal to
print(int(a >= b))  # Greater than or equal to
print(int(a <= b))  # Less than or equal to

# Existing boolean variable as 1/0
print(int(boolean_var)) 
````

An **if statement** allows your program to make decisions. It checks whether a condition is **True** or **False** and executes different code based on the result.

Basic if

Runs code only if the condition is True.

```python
if a > b:
    print("a is greater than b")
```

if-else

Use else when you want your program to do something if the condition is False.

```python
if a < b:
    print("a is less than b")
else:
    print("a is not less than b")
```
if-elif-else

Use elif to check multiple conditions. Python evaluates them from top to bottom and executes the **first** condition that is True.

```python
if a > 20:
    print("a is greater than 20")
elif a > 10:
    print("a is greater than 10")
elif a > 5:
    print("a is greater than 5")
else:
    print("a is 5 or less")
```

**Note:** == checks if two values are equal, while = assigns a value to a variable.
